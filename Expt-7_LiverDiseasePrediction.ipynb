```python
import pandas as pd
data = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis = 1)
data.drop_duplicates(inplace=True)
data
```

```python
data.replace({'gender': {'Male':0, 'Female':1}}, inplace = True)
abs(data.corr()['is_patient']).sort_values(ascending=False)
```

```python
data = data.drop(columns = ['gender','sgpt'])
data
```

```python
X = data.drop(columns = ['is_patient']).values
Y = data['is_patient']

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

from sklearn.model_selection import train_test_split
XTrain, XTest, YTrain, YTest = train_test_split(X,Y, test_size= 0.20, random_state= 2)

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors= 4)
model.fit(XTrain, YTrain)
```

```python
YPred1 = model.predict(XTest)

from sklearn.metrics import accuracy_score
acc = accuracy_score(YTest,YPred1)
print("Accuracy Score of KNN Model :- ", acc*100)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(XTrain, YTrain)
```

```python
YPred2 = log.predict(XTest)

from sklearn.metrics import accuracy_score
acc = accuracy_score(YTest,YPred2)
print("Accuracy Score of Logistic Regression :- ", acc*100)
```
