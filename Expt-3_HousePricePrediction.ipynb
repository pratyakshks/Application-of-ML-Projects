## Training-Test Splitting

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
print("train set rows: ", len(train_set))
print("test set rows: ", len(test_set))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

## Imputing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'median')
imputer.fit(housing)
```

```python
imputer.statistics_.shape
```

```python
imputer.statistics_
```

```python
X = imputer.transform(housing)
```

```python
housing_tr = pd.DataFrame(X, columns = housing.columns)
housing_tr.describe()
```

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
housing_num_tr
```

```python
housing_num_tr.shape
```

## Selecting the Desired model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# model = LinearRegression()


# OR


# model = DecisionTreeRegressor()


# OR


model = RandomForestRegressor()


model.fit(housing_num_tr, housing_labels)
```

```python
some_data = housing.iloc[:5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating the model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better evaluation technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, scoring = 'neg_mean_squared_error')
rsme_scores = np.sqrt(-scores)
```

```python
rsme_scores
```

```python
def print_scores(scores):
    print('Scores: ', scores)
    print('Mean: ', scores.mean())
    print('Standard Deviation: ', scores.std())
```

```python
print_scores(rsme_scores)
```

## Saving the model

```python
from joblib import dump, load
dump(model, 'Price_predictor.joblib')
```

## Testing the model on test data

```python
x_test = strat_test_set.drop('MEDV', axis = 1)
y_test = strat_test_set['MEDV'].copy()
x_test_prepared = my_pipeline.transform(x_test)
final_predictions = model.predict(x_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```
